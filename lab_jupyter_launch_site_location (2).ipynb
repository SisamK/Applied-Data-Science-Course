```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon

# Launch site coordinates (Example: SpaceX launch site)
launch_site_lat = 29.559684888503615
launch_site_lon = -95.0830971930759

# POI types and colors
poi_columns = ['City', 'Railway', 'Highway', 'Coast', 'Coastline']
poi_colors = {'City': 'blue', 'Railway': 'green', 'Highway': 'red', 'Coast': 'purple', 'Coastline': 'orange'}

# Initialize map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=5)

# Function to calculate the distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    from math import radians, sin, cos, sqrt, atan2

    R = 6371.0  # Earth radius in km
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

# Fun icons for each POI type
poi_icons = {
    'City': 'cloud',
    'Railway': 'road',
    'Highway': 'car',
    'Coast': 'anchor',
    'Coastline': 'star'
}

# Iterate through each POI type and add it to the map
for poi_column in poi_columns:
    # This is a placeholder for your actual logic to select the closest POI
    # For simplicity, let's use random coordinates for each type
    poi_lat, poi_lon = launch_site_lat + (0.1 * len(poi_column)), launch_site_lon + (0.1 * len(poi_column))

    # Calculate the distance from the launch site
    distance_poi = calculate_distance(launch_site_lat, launch_site_lon, poi_lat, poi_lon)

    # Create a fun marker for each POI with an icon
    folium.Marker(
        [poi_lat, poi_lon],
        icon=folium.Icon(color=poi_colors[poi_column], icon=poi_icons[poi_column], prefix='fa'),
        popup=f"{poi_column} - Distance: {distance_poi:.2f} km"
    ).add_to(site_map)

    # Create a line from the launch site to the POI
    folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [poi_lat, poi_lon]], color=poi_colors[poi_column], weight=2).add_to(site_map)

# Show the map
site_map
```
